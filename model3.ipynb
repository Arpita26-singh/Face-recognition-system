```python
import os
import cv2
import numpy as np

class WorkingFaceRecognizer:
    def __init__(self):
        self.face_cascade = cv2.CascadeClassifier(
            cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
        self.known_faces = []
        self.known_names = []
    
    def load_images(self):
        print("=== LOADING CONFIRMED IMAGES ===")
        
        # Amar's confirmed working path
        amar_path = r"C:\Users\daara\face_recognition\known_faces\Amar\Amar_1.jpg"
        self._load_face("Amar", amar_path)
        
        # Arpita's confirmed working path
        arpita_path = r"C:\Users\daara\face_recognition\known_faces\Arpita_singh\Arpita_singh.png"
        self._load_face("Arpita", arpita_path)
        
        if len(self.known_faces) != 2:
            print("\nERROR: Not all faces loaded")
            print("Loaded faces:", self.known_names)
            exit()

    def _load_face(self, name, path):
        """Load a single face with verification"""
        if not os.path.exists(path):
            print(f"× File not found: {path}")
            return False
            
        img = cv2.imread(path)
        if img is None:
            print(f"× Could not read image: {path}")
            return False
            
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
        faces = self.face_cascade.detectMultiScale(gray, 1.1, 4, minSize=(100, 100))
        
        if len(faces) == 0:
            print(f"× No face detected in: {path}")
            return False
            
        x,y,w,h = faces[0]
        face = cv2.resize(gray[y:y+h, x:x+w], (200, 200))
        self.known_faces.append(face)
        self.known_names.append(name)
        print(f"✓ Loaded {name} from: {path}")
        return True

    def recognize(self, frame):
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        faces = self.face_cascade.detectMultiScale(gray, 1.1, 4, minSize=(100, 100))
        
        for (x,y,w,h) in faces:
            face_roi = cv2.resize(gray[y:y+h, x:x+w], (200, 200))
            
            best_match = None
            best_score = 0
            
            # Compare with known faces
            for i, known_face in enumerate(self.known_faces):
                score = cv2.matchTemplate(face_roi, known_face, cv2.TM_CCOEFF_NORMED)[0][0]
                if score > best_score:
                    best_score = score
                    best_match = i
            
            # Display results
            if best_match is not None and best_score > 0.4:
                name = self.known_names[best_match]
                color = (0, 255, 0)  # Green
                label = f"{name} ({best_score:.2f})"
            else:
                name = "Unknown"
                color = (0, 0, 255)  # Red
                label = f"Unknown ({best_score:.2f})"
                print(f"No match! Best score: {best_score:.2f}")
            
            cv2.rectangle(frame, (x,y), (x+w,y+h), color, 2)
            cv2.putText(frame, label, (x,y-10), 
                       cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)
        
        return frame

# MAIN EXECUTION
if __name__ == "__main__":
    print("=== WORKING FACE RECOGNITION ===")
    recognizer = WorkingFaceRecognizer()
    recognizer.load_images()
    
    cap = cv2.VideoCapture(0)
    print("\nPress Q to quit...")
    
    while True:
        ret, frame = cap.read()
        if not ret: break
        
        frame = cv2.flip(frame, 1)
        output = recognizer.recognize(frame)
        cv2.imshow('Face Recognition - Working', output)
        
        if cv2.waitKey(1) == ord('q'):
            break
    
    cap.release()
    cv2.destroyAllWindows()
```
