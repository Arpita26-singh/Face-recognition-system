```python
data = [100,200,300,400,500]
index_labels = ['a','b','c','d','e']

s = pd.Series(data, index = index_labels)

print(s)
print(s['b'])
```

```python
data = [5,10,15,20,25]
s = pd.Series(data)

print(s.dtype)
print(s.size)
print(s.index)
print(s.values)
```

```python
data = [10,20,30,40,50]
s = pd.Series(data)

print(s+5)
print(s*2)
print(s.mean())
print(s.max())
print(s.min())
```

```python
data = [5,15,25,35,45]
s = pd.Series(data)

print(s[s>20])
print(s[s%2==0])
```

```python
data = {'a': 100, 'b': 200, 'c': 300, 'd': 400}
s = pd.Series(data)
print(s)
print(s['b'])
```

```python
data = {'a': 10,'b': 20,'c': None,'d': 40}
s = pd.Series(data)

print(s)
print(s.isnull())
print(s.fillna (0))
```

```python
data = [10.20,30,10,20,10]
s = pd.Series(data)

print(s.unique())
print(s.value_counts())
```

```python
data = [1,2,3,4,5]
s = pd.Series(data)

print(s.apply (lambda x: x**2))
print(s.apply (lambda x: x + 10))
```

```python
data = ['1','2','3','4','5']
s = pd.Series(data)
print(s.astype (int))
print(s.astype (float))
print(s.astype (str))
```
