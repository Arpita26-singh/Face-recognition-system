```python
import cv2
import numpy as np
from facenet_pytorch import MTCNN, InceptionResnetV1

# Load FaceNet model
mtcnn = MTCNN()  # Face detector
resnet = InceptionResnetV1(pretrained='vggface2').eval()  # Feature extractor

# Face matching function
def get_face_embedding(frame):
    faces = mtcnn(frame)  # Detect faces
    if faces is not None:
        embeddings = resnet(faces)  # Get 512-D vector
        return embeddings
    return None

# Compare with database
def find_match(embedding, known_embeddings, threshold=0.7):
    distances = np.linalg.norm(known_embeddings - embedding, axis=1)
    min_dist = np.min(distances)
    if min_dist < threshold:
        return "Match Found (Confidence: {:.2f})".format(1 - min_dist)
    return "Unknown Face"
```

```python
import os
import cv2
import numpy as np
import torch
from facenet_pytorch import MTCNN, InceptionResnetV1
from PIL import Image

# Initialize device (use GPU if available)
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
print(f'Using device: {device}')

# Initialize models with optimized settings
mtcnn = MTCNN(
    keep_all=True,
    device=device,
    post_process=False,
    select_largest=False,
    selection_method='largest'
)

resnet = InceptionResnetV1(pretrained='vggface2').eval().to(device)

def load_known_faces(known_faces_dir="known_faces"):
    """Load known faces from directory structure with comprehensive error handling"""
    known_face_encodings = []
    known_face_names = []
    
    # Check if directory exists
    if not os.path.exists(known_faces_dir):
        print(f"Error: Directory '{known_faces_dir}' not found!")
        return torch.tensor([]), []
    
    if not os.path.isdir(known_faces_dir):
        print(f"Error: '{known_faces_dir}' is not a directory!")
        return torch.tensor([]), []
    
    # Get list of subdirectories (people)
    person_dirs = [d for d in os.listdir(known_faces_dir) 
                  if os.path.isdir(os.path.join(known_faces_dir, d))]
    
    if not person_dirs:
        print("Warning: No person directories found in known_faces folder!")
        return torch.tensor([]), []
    
    for person_name in person_dirs:
        person_dir = os.path.join(known_faces_dir, person_name)
        print(f"\nLoading faces for {person_name}...")
        loaded_count = 0
        
        # Get all image files in person's directory
        image_files = [f for f in os.listdir(person_dir) 
                      if f.lower().endswith(('.png', '.jpg', '.jpeg'))]
        
        if not image_files:
            print(f"  Warning: No images found for {person_name}")
            continue
            
        for img_file in image_files:
            img_path = os.path.join(person_dir, img_file)
            try:
                # Open and convert image
                img = Image.open(img_path).convert('RGB')
                
                # Detect face
                face = mtcnn(img)
                
                if face is None:
                    print(f"   No face detected in {img_file}")
                    continue
                    
                # Get embedding
                embedding = resnet(face.unsqueeze(0).to(device))
                known_face_encodings.append(embedding.detach().cpu())
                known_face_names.append(person_name)
                loaded_count += 1
                print(f"   Loaded {img_file}")
                
            except Exception as e:
                print(f"   Error processing {img_file}: {str(e)}")
        
        print(f"  Successfully loaded {loaded_count}/{len(image_files)} images for {person_name}")
    
    if not known_face_encodings:
        print("\nWarning: No faces were successfully loaded from any images!")
        return torch.tensor([]), []
    
    try:
        # Stack all embeddings into single tensor
        stacked_encodings = torch.stack(known_face_encodings)
        print(f"\nSuccessfully loaded {len(known_face_names)} reference faces")
        return stacked_encodings, known_face_names
    except Exception as e:
        print(f"\nError stacking face encodings: {str(e)}")
        return torch.tensor([]), []

def recognize_faces(frame, threshold=0.7):
    """Recognize faces in a video frame with robust error handling"""
    if frame is None or frame.size == 0:
        return []
    
    try:
        # Convert to PIL Image
        img = Image.fromarray(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
        
        # Detect all faces
        boxes, probs = mtcnn.detect(img)
        
        recognized = []
        
        if boxes is not None:
            for i, (box, prob) in enumerate(zip(boxes, probs)):
                if prob < 0.9:  # Skip low confidence detections
                    continue
                    
                # Extract face
                face = mtcnn.extract(img, [box], None)
                
                if face is not None:
                    # Get embedding
                    embedding = resnet(face.to(device)).detach().cpu()
                    
                    # Only compare if we have known faces
                    if known_face_encodings.numel() > 0:
                        # Calculate distances to known faces
                        distances = torch.norm(known_face_encodings - embedding, dim=1)
                        min_distance, best_match_idx = torch.min(distances, 0)
                        
                        if min_distance < threshold:
                            name = known_face_names[best_match_idx]
                            confidence = 1 - (min_distance / threshold)
                        else:
                            name = "Unknown"
                            confidence = 0
                    else:
                        name = "Unknown (no reference)"
                        confidence = 0
                        min_distance = float('inf')
                    
                    recognized.append({
                        'box': box,
                        'name': name,
                        'confidence': confidence,
                        'distance': min_distance.item()
                    })
        
        return recognized
    
    except Exception as e:
        print(f"Error in face recognition: {str(e)}")
        return []

def display_results(frame, recognized_faces):
    """Draw recognition results on the frame"""
    if frame is None or frame.size == 0:
        return frame
    
    for face in recognized_faces:
        box = face['box']
        name = face['name']
        confidence = face['confidence']
        
        # Convert box coordinates to integers
        x1, y1, x2, y2 = map(int, box)
        
        # Draw bounding box
        color = (0, 255, 0) if name != "Unknown" and not name.startswith("Unknown (") else (0, 0, 255)
        cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
        
        # Draw label background
        label = f"{name} ({confidence:.2f})" if confidence > 0 else name
        (label_width, label_height), _ = cv2.getTextSize(label, cv2.FONT_HERSHEY_SIMPLEX, 0.8, 2)
        cv2.rectangle(frame, (x1, y1 - label_height - 10), (x1 + label_width, y1), color, -1)
        
        # Draw label text
        cv2.putText(frame, label, (x1, y1 - 10), 
                    cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    
    return frame

def real_time_recognition():
    """Run real-time face recognition with proper resource management"""
    cap = None
    try:
        cap = cv2.VideoCapture(0)
        if not cap.isOpened():
            print("Error: Could not open video capture device")
            return
        
        print("\nStarting real-time recognition. Press 'q' to quit...")
        
        while True:
            ret, frame = cap.read()
            if not ret:
                print("Warning: Could not read frame from camera")
                break
            
            # Mirror the frame for more natural interaction
            frame = cv2.flip(frame, 1)
            
            # Recognize faces
            recognized_faces = recognize_faces(frame)
            
            # Display results
            frame = display_results(frame, recognized_faces)
            
            # Show frame
            cv2.imshow('Face Recognition', frame)
            
            # Exit on 'q'
            if cv2.waitKey(1) & 0xFF == ord('q'):
                break
                
    except Exception as e:
        print(f"Error in real-time recognition: {str(e)}")
    finally:
        if cap is not None:
            cap.release()
        cv2.destroyAllWindows()

if __name__ == "__main__":
    print("Initializing face recognition system...")
    
    # Load known faces
    known_face_encodings, known_face_names = load_known_faces()
    
    # Only proceed if we loaded at least one face
    if known_face_encodings.numel() > 0:
        real_time_recognition()
    else:
        print("Cannot start recognition without any known faces. Please add images to the known_faces directory.")
```
